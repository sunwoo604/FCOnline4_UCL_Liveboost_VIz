```python
import pandas as pd
import numpy as np
```

```python
boost16 = pd.read_csv('boost16.csv')
boost8 = pd.read_csv('boost8.csv')
boost4 = pd.read_csv('boost4.csv')
```

```python
boosts = pd.concat([boost16,boost8,boost4])
```

```python
boosts = boosts[['name','team']]
```

```python
boosts['name'].unique()
```

```python
temp = boosts.set_index('name')
boosts = pd.DataFrame(boosts['name'].unique())
```

```python
def team(x):
    if len(temp.loc[x])==1:
        return temp.loc[x]['team']
    return temp.loc[x].iloc[0]['team']
```

```python
boosts['team'] = boosts[0].apply(team)
```

```python
boosts.columns=['player','team']
```

```python
boosts.to_csv('player_team.csv',index=False)
```
